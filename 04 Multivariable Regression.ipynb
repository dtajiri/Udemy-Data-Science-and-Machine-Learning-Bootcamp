# Training & Test Dataset Split

```python
prices = data['PRICE']
features = data.drop('PRICE', axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state = 10)

# % of training set
len(X_train)/len(features)
```

```python
# % of test data set
X_test.shape[0]/features.shape[0]
```

# Multivariable Regression

```python
regr = LinearRegression()
regr.fit(X_train, y_train)

#Challenge: print out r-square for training and testing dataset
print('Traininig data r-squared:', regr.score(X_train,y_train))
print('Test data r-squared:', regr.score(X_test, y_test))
#its normal for performance to lower for the test data, but it shows how our model would work in the real world

print('Intercept', regr.intercept_)
pd.DataFrame(data = regr.coef_, index= X_train.columns, columns=['coef'])
```

## p values & Evaluating Coefficients

```python
X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const) #OLS --> Ordinary Least Squares
results = model.fit()

#results.params
#results.pvalues
pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues,3)})
#according to p-values, the INDUS and AGE features aren't very significant
```

## Testing for Multicollinearity
$$ TAX = \alpha_0 + \alpha_1 RM + \alpha_2 NOX + ... + \alpha_{12} LSTAT$$
$$ VIF_{TAX} = \frac{1}{(1 - R_{TAX} ^2)}$$

```python
variance_inflation_factor(exog=X_incl_const.values, exog_idx = 1)
```

```python
#Challenge: print out the number of columns in X_incl_const
print(X_incl_const.values.shape[1])
print(len(X_incl_const.columns))
```

```python
# Challenge: write a for loop that prints out all the VIFs for all the features
for i in range(len(X_incl_const.columns)):
    print(variance_inflation_factor(exog=X_incl_const.values, exog_idx = i))
```

```python
vif_all = [variance_inflation_factor(exog=X_incl_const.values, exog_idx = i)
           for i in range(X_incl_const.shape[1])]
pd.DataFrame({'coef_name': X_incl_const.columns, 'vif': np.round(vif_all,2)})
```

# Model Simplification  & BIC

```python
# Original model with log prices and all features

X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const) #OLS --> Ordinary Least Squares
results = model.fit()

orig_coef = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues,3)})

# Challenge: find and check official docs for results object and print out BIC & r-squared
print('R-squared: ', results.rsquared)
print('BIC:', results.bic)
```

```python
# reduced model #1 excluding INDUS
X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS'], axis = 1)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

coef_minus_indus = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues,3)})

# Challenge: find and check official docs for results object and print out BIC & r-squared
print('R-squared: ', results.rsquared)
print('BIC:', results.bic)

# results encouraging because the BIC is lower and the goodness of fit stays relatively the same
```

```python
# reduced model #2 excluding INDUS and AGE
X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS', 'AGE'], axis = 1)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

reduced_coef = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues,3)})

print('R-squared: ', results.rsquared)
print('BIC:', results.bic)

# results show we can remove INDUS and AGE without much loss. This simplifies our model
```

```python
frames = [orig_coef, coef_minus_indus, reduced_coef]
pd.concat(frames,axis=1)
```

# Residuals and Residual Plots

```python
# Modified model: transformed (using log prices) & simplified (dropping two features)
prices = np.log(data['PRICE']) #use log prices
features = data.drop(['PRICE', 'INDUS', 'AGE'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state = 10)

# Using statsmodel

X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

# Residuals
#residuals = y_train - results.fittedvalues
#results.resid

#Graph of Actual vs. Predicted Prices
corr = round(y_train.corr(results.fittedvalues),2)

plt.style.use('seaborn-whitegrid')

plt.plot(y_train,y_train, color='cyan')
plt.scatter(x=y_train, y=results.fittedvalues, c = 'navy', alpha = .5)

plt.xlabel('Actual log prices $y_i$', fontsize=14)
plt.ylabel('Predicted log prices $\hat y_i$', fontsize=14)
plt.title(f'Actual vs Predicted log prices: $y_i$ vs $\hat y_i$ (Corr = {corr})', fontsize = 17)

plt.show()

plt.plot(np.e**y_train,np.e**y_train, color='cyan')
plt.scatter(x=np.e**y_train, y=np.e**results.fittedvalues, c = 'blue', alpha = .5)

plt.xlabel('Actual prices 000s $y_i$', fontsize=14)
plt.ylabel('Predicted prices 000s $\hat y_i$', fontsize=14)
plt.title(f'Actual vs Predicted prices: $y_i$ vs $\hat y_i$ (Corr = {corr})', fontsize = 17)

plt.show()

# Residuals vs Predicted values

plt.scatter(x=results.fittedvalues, y=results.resid, c = 'purple', alpha = .5)

plt.xlabel('Predicted log prices $\hat y_i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted Values', fontsize = 17)

plt.show()

#Mean Squared Error & R-Squared
reduced_log_mse = round(results.mse_resid,3)
reduced_log_rsquared = round(results.rsquared, 3)
```

```python
# Distributtion of Residuals (log prices) - checking for normality
resid_mean = round(results.resid.mean(),3)
resid_skew = round(results.resid.skew(),3)

sns.histplot(results.resid, color='navy', kde = True)
plt.title(f'Log price model: residuals Skew({resid_skew}) Mean({resid_mean})')
plt.show()
```

```python
# Challenge: Using the original model with all the features and normal prices generate:
# Plot of actual vs predicted prices (incl. correlation) using a different color
# Plot of residuals vs. predicted prices
# Plot of distribution of residuals (incl. skew)
# Analyze the results
prices = data['PRICE']
features = data.drop(['PRICE'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state = 10)

X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const)
orig_results = model.fit()

#Graph of Actual vs. Predicted Prices
corr = round(y_train.corr(orig_results.fittedvalues),2)

plt.plot(y_train,y_train, color='cyan')
plt.scatter(x=y_train, y=orig_results.fittedvalues, c = 'navy', alpha = .5)

plt.xlabel('Actual prices $y_i$', fontsize=14)
plt.ylabel('Predicted  prices $\hat y_i$', fontsize=14)
plt.title(f'Actual vs Predicted prices: $y_i$ vs $\hat y_i$ (Corr = {corr})', fontsize = 17)

plt.show()

# Residuals vs Predicted values

plt.scatter(x=orig_results.fittedvalues, y=orig_results.resid, c = 'purple', alpha = .5)

plt.xlabel('Predicted prices $\hat y_i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted Values', fontsize = 17)

plt.show()

# Distributtion of Residuals (log prices) - checking for normality
orig_resid_mean = round(orig_results.resid.mean(),3)
orig_resid_skew = round(orig_results.resid.skew(),3)

sns.histplot(orig_results.resid, color='navy', kde = True)
plt.title(f'Price model: residuals Skew({orig_resid_skew}) Mean({orig_resid_mean})')
plt.show()

#Mean Squared Error & R-Squared
full_normal_mse = round(orig_results.mse_resid,3)
full_normal_rsquared = round(orig_results.rsquared, 3)
```

```python
pd.DataFrame({'R-Squared': [reduced_log_rsquared, full_normal_rsquared], 
              'MSE' : [reduced_log_mse, full_normal_mse], 
              'RMSE': np.sqrt([reduced_log_mse, full_normal_mse])
             }, index=['Reduced Log Model', 'Full Normal Price Model'],)

# scale of the MSE is dependent on the data.
# lower values of MSE means better fit, but one is in log and one is normal
```

```python
# Challenge: Our estimate for a house price is $30,000. Calculate the upper and lower bound
# for a 95% prediction interval using the reduced log model

print('1 std in log prices is: ', np.sqrt(reduced_log_mse))
print('2 std in log prices is: ', 2*np.sqrt(reduced_log_mse))
print('Upper bound in log is: ', np.log(30) + 2 *np.sqrt(reduced_log_mse))
print('Upper bound in normal prices is: ', np.e**(np.log(30) + 2 *np.sqrt(reduced_log_mse)) *1000)
print('Lower bound in log is: ', np.log(30) - 2 *np.sqrt(reduced_log_mse))
print('Lower bound in normal prices is: ', np.e**(np.log(30) - 2 *np.sqrt(reduced_log_mse))*1000)
```
